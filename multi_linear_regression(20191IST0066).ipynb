Use only the following features to predict the players sold price.

my X_features are independent variables used to find my dependent variable(Y)

```python
X_features=['AGE', 'COUNTRY', 'PLAYING ROLE', 'T-RUNS', 'T-WKTS', 'ODI-RUNS-S', 'ODI-SR-B', 'ODI-WKTS', 'ODI-SR-BL','CAPTAINCY EXP', 'RUNS-S', 'HS', 'AVE', 'SR-B','SIXERS', 'RUNS-C', 'WKTS', 'AVE-BL', 'ECON', 'SR-BL']
```

Some of these features are categorical. So encode them using one-hot encoding. A categorical feature with k values is encoded using (k-1) binary/dummy features.

```python
categorical_features=['AGE', 'COUNTRY','PLAYING ROLE','CAPTAINCY EXP']
encoded_df = pd.get_dummies(df[X_features], columns=categorical_features, drop_first = True)
#print the columns in the encoded dataframe
X_features = encoded_df.columns
print(X_features)
```

Add the constant column to estimate Beta 0 using statsmodel library. Split the dataset into Train and Validation Sets

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
X = sm.add_constant(encoded_df)
Y = df['SOLD PRICE']

train_X, test_X, train_y, test_y = train_test_split(X,Y,train_size = 0.8, random_state = 42)
```

Build the model on training set

```python
ipl_mlr_model1 = sm.OLS(train_y,train_X).fit()
ipl_mlr_model1.summary2()
```

Validate this model on the test data. Find the regression parameters, r2_score and RMSE of this model1

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

pred_y = ipl_mlr_model1.predict(test_X)
print("The r2_score of the model is", np.abs(r2_score(test_y, pred_y)))
print("The RMSE of the trained model is", np.sqrt(mean_squared_error(test_y, pred_y)))
print("The model parameters are")
print(ipl_mlr_model1.params)
```
